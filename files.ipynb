### Кодировки текстовых файлов

**Что такое кодировка?**
- Кодировка - это таблица соответствия между символами и числами
- Компьютер хранит только числа, а кодировка говорит, какому числу какой символ соответствует

**Популярные кодировки:**

1. **UTF-8** (современный стандарт)
   - Поддерживает все языки мира
   - Совместима с ASCII
   - Использует от 1 до 4 байт на символ

2. **Windows-1251** (CP-1251)
   - Для кириллицы (русский, украинский и др.)
   - Используется в Windows
   - 1 байт на символ

3. **KOI8-R**
   - Для русского языка
   - Популярна в Unix-системах
   - 1 байт на символ

4. **ASCII**
   - Только английские буквы, цифры и основные символы
   - 1 байт на символ
   - Не поддерживает кириллицу!

**Проблема с кодировками:**

- "Hello" → "╨ÿ╨╡╨╗╨╗╨╛" (Windows-1251 прочитан как UTF-8)

### Работа с файлами

**Основные режимы открытия файлов:**

1. **`'r'` - Read (чтение)**
   - Открывает файл только для чтения
   - Файл должен существовать

2. **`'w'` - Write (запись)**
   - Открывает файл для записи
   - Если файл существует - перезаписывает его
   - Если файла нет - создает новый
   - **Удаляет старое содержимое**

3. **`'a'` - Append (добавление)**
   - Открывает файл для добавления данных
   - Если файл существует - добавляет в конец
   - Если файла нет - создает новый
   - Сохраняет старое содержимое

4. **`'r+'` - Read and Write (чтение и запись)**
   - Открывает файл для чтения и записи
   - Файл должен существовать

### Синтаксис

**Способ 1: Классический (нужно закрывать вручную)**
```python
file = open('filename.txt', 'r', encoding='utf-8')
content = file.read()
file.close()  # ВАЖНО: всегда закрывать файл!
```

**Способ 2: Конструкция with (рекомендуется)**
```python
with open('filename.txt', 'r', encoding='utf-8') as file:
    content = file.read()
# Файл автоматически закрывается после выхода из блока with
```

### Методы чтения файлов:

1. **`read()` - чтение всего файла**

````python
with open('file.txt', 'r', encoding='utf-8') as file:
    entire_content = file.read()
    print(entire_content)
````

2. **`readline()` - чтение одной строки**

````python
with open('file.txt', 'r', encoding='utf-8') as file:
    first_line = file.readline()
    second_line = file.readline()
    print(first_line)
    print(second_line)
````

3. **`readlines()` - чтение всех строк в список**

````python
with open('file.txt', 'r', encoding='utf-8') as file:
    all_lines = file.readlines()
    for line in all_lines:
        print(line.strip())  # strip() убирает \n в конце
````

4. **Чтение файла построчно через цикл**

````python
with open('file.txt', 'r', encoding='utf-8') as file:
    for line in file:
        print(line.strip())
````

### Методы записи

1. **Запись строки**

````python
with open('output.txt', 'w', encoding='utf-8') as file:
    file.write("Привет, мир!\n")
    file.write("Это вторая строка\n")
````

2. **Запись списка строк**

````python
lines = ["Первая строка\n", "Вторая строка\n", "Третья строка\n"]

with open('output.txt', 'w', encoding='utf-8') as file:
    file.writelines(lines)
````

3. **Добавление в конец файла**

````python
with open('output.txt', 'a', encoding='utf-8') as file:
    file.write("Эта строка добавится в конец\n")
````
