## Анализ и статистика

```python
# Статистика по странам
country_stats = df['Country'].value_counts()
print("Топ-5 стран по количеству клиентов:")
print(country_stats.head())

# Анализ по дате подписки
df['Subscription Year'] = pd.to_datetime(df['Subscription Date']).dt.year
yearly_subscriptions = df['Subscription Year'].value_counts().sort_index()

print("\nПодписки по годам:")
print(yearly_subscriptions)
```

- **dt** - работа с датами в pandas: `df['дата'].dt.year`, `.dt.month`, `.dt.day` для извлечения частей даты

- **datetime** - модуль для работы с датами: `pd.to_datetime(df['столбец'])` преобразует текст в даты, `datetime.now()` - текущая дата-время

- **value_counts** - подсчет уникальных значений: `df['столбец'].value_counts()` - возвращает частоту каждого значения, `normalize=True` для долей

- **sort_index** - сортировка по индексу: `df.sort_index()` - по возрастанию, `ascending=False` - по убыванию. После groupby или для изменения порядка строк

## Создание и сохранение новых таблиц

```python
# Создаем таблицу с аналитикой по странам
country_analysis = df.groupby('Country').agg({
    'First Name': 'count',
    'Name Length': 'mean',
    'Subscription Date': 'min'
}).rename(columns={
    'First Name': 'Total Customers',
    'Name Length': 'Average Name Length',
    'Subscription Date': 'First Subscription'
}).round(2)

print("Аналитика по странам:")
print(country_analysis.head())

# Создаем таблицу контактов
contacts_table = df[['Full Name', 'Email', 'Phone 1 Clean', 'Country', 'Company']]
contacts_table = contacts_table.rename(columns={'Phone 1 Clean': 'Phone'})

print("\nТаблица контактов:")
print(contacts_table.head())
```

- **Groupby** - группировка данных по столбцам для агрегации (сумма, среднее, количество). Пример: `df.groupby('страна')['продажи'].sum()`

- **agg** - применяет несколько агрегатных функций сразу: `df.agg({'столбец': ['mean', 'sum', 'count']})` или `df.groupby('x').agg({'y': 'sum', 'z': 'mean'})`

## Туториал: Обработка текстовых данных с Pandas

### Установка библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from collections import Counter
```

### Создание датасета

```python
# Создаем реалистичный датасет отзывов о продуктах
data = {
    'review_text': [
        "This product is absolutely amazing! The quality exceeded my expectations.",
        "Terrible customer service and the product arrived broken. Very disappointed.",
        "It's okay, nothing special. Does the job but I expected more features.",
        "The delivery was fast but the product quality is mediocre at best.",
        "Outstanding value for money! Will definitely buy again from this brand.",
        "Worst purchase ever. The item stopped working after 2 days.",
        "Good product with excellent features. The packaging was also very nice.",
        "Average product, average service. Nothing to complain about.",
        "The customer support team was very helpful and resolved my issue quickly!",
        "Product is good but the shipping took too long. Overall satisfied."
    ],
    'rating': [5, 1, 3, 2, 5, 1, 4, 3, 4, 3],
    'product_category': ['electronics', 'electronics', 'home', 'home', 'electronics',
                        'clothing', 'electronics', 'home', 'electronics', 'clothing'],
    'review_length': [58, 52, 43, 47, 51, 32, 48, 38, 62, 45]
}

df = pd.DataFrame(data)
print(f"Размер датасета: {df.shape}")
df.head()
```

### Операции и преобразования

```python
# Просмотр информации о датасете
df.info()
```

```python
df.describe()
```

```python
# Анализ распределения рейтингов: продвинутое использование
df['rating'].value_counts().sort_index()
```

```python
# Приведение к нижнему регистру
df['cleaned_text'] = df['review_text'].str.lower()

# Удаление пунктуации и специальных символов
df['cleaned_text'] = df['cleaned_text'].str.replace(r'[^\w\s]', '', regex=True)

# Удаление лишних пробелов
df['cleaned_text'] = df['cleaned_text'].str.strip()
df['cleaned_text'] = df['cleaned_text'].str.replace(r'\s+', ' ', regex=True)

# Показываем результаты
df[['review_text', 'cleaned_text']].head()
```

### Создание новых признаков

```python
# Длина текста
df['text_length'] = df['cleaned_text'].str.len()

# Количество слов
df['word_count'] = df['cleaned_text'].str.split().str.len()

# Определение тональности по ключевым словам (простой подход)
def simple_sentiment(text):
    positive_words = ['amazing', 'outstanding', 'excellent', 'good', 'great', 'helpful', 'satisfied']
    negative_words = ['terrible', 'broken', 'disappointed', 'worst', 'mediocre']

    text_lower = text.lower()
    positive_score = sum(1 for word in positive_words if word in text_lower)
    negative_score = sum(1 for word in negative_words if word in text_lower)

    if positive_score > negative_score:
        return 'positive'
    elif negative_score > positive_score:
        return 'negative'
    else:
        return 'neutral'

df['simple_sentiment'] = df['cleaned_text'].apply(simple_sentiment)

df[['cleaned_text', 'text_length', 'word_count', 'simple_sentiment']].head()
```

### Визуализация

```python
# Настройка стиля графиков
sns.set_theme(style="whitegrid")
plt.figure(figsize=(15, 10))

# 1. Распределение рейтингов
plt.subplot(2, 3, 1)
df['rating'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Распределение рейтингов')
plt.xlabel('Рейтинг')
plt.ylabel('Количество')

# 2. Распределение длины текста
plt.subplot(2, 3, 2)
plt.hist(df['text_length'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Распределение длины текста')
plt.xlabel('Длина текста')
plt.ylabel('Частота')

# 3. Распределение количества слов
plt.subplot(2, 3, 3)
sns.boxplot(x='rating', y='word_count', data=df)
plt.title('Количество слов по рейтингам')
plt.xlabel('Рейтинг')

# 4. Сентимент анализ
plt.subplot(2, 3, 4)
df['simple_sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral', 'lightyellow'])
plt.title('Распределение тональности')

# 5. Распределение по категориям
plt.subplot(2, 3, 5)
df['product_category'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Распределение по категориям')
plt.xlabel('Категория')
plt.ylabel('Количество')

plt.tight_layout()
plt.show()
```
